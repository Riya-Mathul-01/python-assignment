```python
import random

numbers = [random.randint(1, 100) for i in range(10)]

print("Random Numbers:", numbers)

print("Even Numbers:")
for num in numbers:
    if num % 2 == 0:
        print(num)
```

```python
text = "Python is very easy"

print("Original String:")
print(text)

print("\nReversed String:")
print(text[::-1])

print("\nWords Reversed:")
print(" ".join(text.split()[::-1]))
```
